# About the Dataset
### The Dataset contain 1030 books obtainted from scraping Amazon books related to data science, statics ,data analysis ,python ,deep learing and machine learning 

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python

df = pd.read_csv(r"C:\Users\HP\Downloads\final_book_dataset_kaggle.csv")

# Display first 5 rows
df.head()
```

```python
# Shape of dataset
print("Rows and Columns:", df.shape)

# Column names
print("\nColumn Names:")
print(df.columns.tolist())
```

```python
# Dataset information (datatypes + null counts)
print("Dataset Info:")
df.info()

# Statistical summary
print("\nStatistical Summary:")
df.describe()
```

```python
# Check how many missing values exist in each column
print("Missing Values:")
print(df.isnull().sum())

# Percentage of missing values
print("\nMissing %:")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Remove rows with null values
df = df.dropna()

# Verify no nulls remain
print("Missing values after cleaning:")
print(df.isnull().sum())
print("\nNew shape:", df.shape)
```

```python
# Check duplicate rows
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Verify
print("After removing duplicates:", df.duplicated().sum())
print("Final shape:", df.shape)
```

```python
# Check data types of all columns
print(df.dtypes)

# Ensure numeric columns are correct type
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['pages'] = pd.to_numeric(df['pages'], errors='coerce')
df['avg_reviews'] = pd.to_numeric(df['avg_reviews'], errors='coerce')
df['n_reviews'] = pd.to_numeric(df['n_reviews'], errors='coerce')

print("\nData types after conversion:")
print(df.dtypes)
```

```python
# Total books
print("Total Books:", df['title'].count())

# Average review rating
print("Average Rating:", df['avg_reviews'].mean().round(2))

# Average price
print("Average Price: $", df['price'].mean().round(2))

# Most common rating
print("\nRating value counts:")
print(df['avg_reviews'].value_counts().head(10))
```

# Top 10 most expensive books📔

```python
top_expensive = df.sort_values(by='price', ascending=False)
print(top_expensive[['title', 'price']].head(10).to_string())
```

# Top 10 highest rated books📔

```python
highest_rated = df.sort_values(by='avg_reviews', ascending=False)
print(highest_rated[['title', 'avg_reviews', 'n_reviews']].head(10).to_string())
```

```python
# Average price grouped by rating
avg_price_by_rating = df.groupby('avg_reviews')['price'].mean().round(2)
print("Average Price by Rating:")
print(avg_price_by_rating)

# Average pages grouped by rating
avg_pages_by_rating = df.groupby('avg_reviews')['pages'].mean().round(0)
print("\nAverage Pages by Rating:")
print(avg_pages_by_rating)
```

```python
# Distribution of average review ratings
plt.figure(figsize=(10, 5))
sns.histplot(df['avg_reviews'], bins=20, color='steelblue', kde=True)
plt.title('Distribution of Book Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

# Bar chart of top 10 most expensive books📔

```python
# Bar chart of top 10 most expensive books
plt.figure(figsize=(12, 6))
expensive_books = df.sort_values(by='price', ascending=False).head(10)
sns.barplot(x='price', y='title', data=expensive_books, palette='Blues_d')
plt.title('Top 10 Most Expensive Books')
plt.xlabel('Price ($)')
plt.ylabel('Book Title')
plt.tight_layout()
plt.show()
```

```python
# Countplot of rating values
plt.figure(figsize=(12, 5))
sns.countplot(x='avg_reviews', data=df, palette='viridis')
plt.title('Book Ratings Count')
plt.xlabel('Rating')
plt.ylabel('Total Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Price distribution with boxplot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['price'], bins=30, ax=axes[0], color='coral', kde=True)
axes[0].set_title('Price Distribution')
axes[0].set_xlabel('Price ($)')

sns.boxplot(y=df['price'], ax=axes[1], color='coral')
axes[1].set_title('Price Boxplot (Outliers)')
axes[1].set_ylabel('Price ($)')

plt.tight_layout()
plt.show()
```

```python
# Pages distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['pages'], bins=30, color='mediumseagreen', kde=True)
plt.title('Distribution of Book Pages')
plt.xlabel('Number of Pages')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: price vs average rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='avg_reviews', data=df,
               alpha=0.6, color='steelblue')
plt.title('Price vs Average Rating')
plt.xlabel('Price ($)')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# Average proportion of each star rating across all books
star_means = df[['star5','star4','star3','star2','star1']].mean()

plt.figure(figsize=(8, 5))
star_means.plot(kind='bar', color=['#2ecc71','#27ae60','#f39c12','#e67e22','#e74c3c'])
plt.title('Average Star Rating Distribution')
plt.xlabel('Star Rating')
plt.ylabel('Average Proportion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap for all numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10, 7))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f',
           cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Save the cleaned dataset
df.to_csv('Cleaned_Amazon_Books.csv', index=False)
print("Cleaned dataset saved successfully!")
print(f"Final dataset shape: {df.shape}")
print("Project completed successfully!")
```
